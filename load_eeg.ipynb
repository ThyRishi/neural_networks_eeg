```python
import mne
%matplotlib notebook
import matplotlib.pyplot as plt
import numpy as np
```

```python
def reshape_data(dataset):
    epoched_data = dataset.get_data()
    reshaped_data = np.concatenate([np.concatenate([epoched_data[i:i + 1, :, j:j + 100] 
                                                    for j in np.arange(0, epoched_data.shape[2], 100)], axis=0)
                                    for i in range(epoched_data.shape[0])], axis=0)
    dataset = mne.EpochsArray(reshaped_data, dataset.info)
    return dataset
```

```python
# Specify the path to the files
data_path = 'propofol_data'
```

```python
conscious = mne.read_epochs_eeglab(f'{data_path}/coad_01_1020.set')
```

```python
conscious = reshape_data(conscious)
```

```python
conscious.plot(n_epochs=10)
plt.show()
```

```python
unconscious = mne.read_epochs_eeglab(f'{data_path}/coad_03_1020.set')
```

```python
unconscious = reshape_data(unconscious)
```

```python
unconscious.plot(n_epochs=10)
plt.show()
```
